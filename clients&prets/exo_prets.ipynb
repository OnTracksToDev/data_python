```python
import numpy as np
import pandas as pd
```

```python
prets = pd.read_csv("prets.txt")
prets.head()
```

```python
prets['taux_endettement'] = round((prets['remboursement'] / prets['revenu']) * 100, 2)
```

```python
prets[['identifiant', 'revenu', 'remboursement', 'taux_endettement']].head()
```

```python
prets= prets.rename(columns={'taux':'taux_interets'})
prets
```

```python
prets['cout_total'] = prets['remboursement'] * prets['duree']
prets['cout_total']
```

```python
prets['benefices'] = round((prets['cout_total'] * prets['taux_interets']/100)/(24), 2)
prets
```

```python
prets.sort_values('benefices', ascending= False).head()
```
